```python
#有一個name.txt的檔案,裏面有299個名字,每行一個名字
#請寫一個程式,讀取這個檔案,並且將讀出每一個名字
#並且將讀出的名字存入一個list中,然後印出這個list
#如果檔案不存在,請印出"檔案不存在"
# 嘗試讀取 names.txt 檔案，將每一行的名字存入 names list，並印出 list
# 如果檔案不存在，則印出 "檔案不存在"


def read_names(file_path):
    names = []  # 初始化一個空的 list 
    
    try:
        # 使用 with open 確保檔案會自動關閉
        with open(file_path, mode='r', encoding='utf-8') as file:
            for line in file:
                # 使用 strip() 去除每行末尾的換行符號 \n
                name = line.strip()
                # 使用 list 的 append 方法將名字加入 
                names.append(name)
        
        # 讀取完成後印出整個 list 
        print(names)
        return names

    except FileNotFoundError:
        # 如果檔案不存在，印出指定訊息 
        print("檔案不存在")
        return None
    except Exception as e:
        print(f"發生其他錯誤：{e}")
        return None

if __name__ == "__main__":
    # 執行程式碼
    read_names('names.txt')
```
