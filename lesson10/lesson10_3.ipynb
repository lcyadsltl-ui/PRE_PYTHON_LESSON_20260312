```python
my_list:list = [i for i in range(10)]
print(type(my_list))
#list的實體方法
my_list.append(10)
print(my_list)
```

```python
#建立my_list常用的實體方法
my_list.clear()  # 清空列表
print(my_list)
my_list.append(10)  # 在列表末尾添加元素
print(my_list)
my_list.extend([11, 12])  # 在列表末尾添加多個元素
print(my_list)
my_list.insert(0, -1)  # 在指定位置插入元素
print(my_list)
my_list.remove(10)  # 移除列表中的第一個匹配項
print(my_list)
print(my_list.pop())  # 移除並返回列表中的最後一個元素
print(my_list)
```
