```python
#複合指定運算子

total = 0
value = float(input("請輸入第1個值:"))
total += value

value = float(input("請輸入第2個值:"))
total += value

value = float(input("請輸入第3個值:"))
total += value

print("3數的加總是:", total)
```

```python
#直角三角形斜邊計算
import math
a = float(input("請輸入a="))
b = float(input("請輸入b="))
c = math.sqrt(a ** 2 + b ** 2)

print("a=", a)
print("b=", b)
print("斜邊為", c)
```

```python
import math
import random
import datetime

# 隨機產生一個 100 到 500 的房價（假設單位是萬）
price = random.randint(100, 500)
# 計算這個房價的平方根
root_value = math.sqrt(price)
# 取得現在時間
now = datetime.datetime.now()

print(f"現在時間：{now}")
print(f"隨機房價：{price} 萬，其平方根為：{root_value:.2f}")
```

```python
# 練習使用內建函數 sum() 和 len()
house_prices = [2500, 3200, 1800] # 萬元
total = sum(house_prices)
average = total / len(house_prices)

print(f"總價是：{total} 萬元")
print(f"平均價是：{average:.1f} 萬元")
```
