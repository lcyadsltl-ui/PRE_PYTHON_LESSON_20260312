```python
score=float(input("學生成績:"))
print("學生成績:",score)
if score>=60:
    print("及格")
    print("恭喜你通過了！")
else:
    print("不及格")
    print("很遺憾，你需要繼續努力。")
```

```python
score=float(input("學生總成績(0-300):"))
print("學生總成績:",score)
is_add=input("學生是否符合加分: Y/N")
if is_add=="Y":
    score+=10
    print("加分後成績:",score)
else:
    print("沒有加分，成績保持不變:",score)
```
