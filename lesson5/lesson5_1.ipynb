```python
score=float(input("學生總成績(0-300):"))
print("學生總成績:",score)
if not(score<0 or score>300):
    is_add=input("學生是否符合加分: Y/N")
    if is_add=="Y":
        score+=10
        if score>300:
            score=300
        print("加分後成績:",score)
    else:
        print("沒有加分，成績保持不變:",score)
else:    print("成績在不正常範圍內，重新輸入:",score)
print("程式結束")
```
