```python
try:
    score=float(input("學生總成績(0-300):"))
except ValueError as error:
    print("輸入錯誤，請輸入數字:",error)
except Exception as e:
    print("發生未知錯誤:",e)
print("學生總成績:",score)
if not(score<0 or score>300):
    is_add=input("學生是否符合加分: Y/N")
    if is_add=="Y":
        score+=10
        if score>300:
            score=300
        print("加分後成績:",score)
    else:
        print("沒有加分，成績保持不變:",score)
else:    print("成績在不正常範圍內，重新輸入:",score)
print("程式結束")
```

```python
try:
    身高 = float(input("請輸入您的身高"))
    print("您的身高為:", 身高)
except ValueError as error:
    print("身高輸入錯誤，請輸入數字:",error)
except Exception as e:
    print("發生未知錯誤:",e)
else:
    try:
        體重 = float(input("請輸入您的體重"))
        print("您的體重為:", 體重)
    except ValueError as error:
        print("體重輸入錯誤，請輸入數字:",error)
    except Exception as e:
        print("發生未知錯誤:",e)
    else:
        BMI = 體重 / (身高/100)**2
        print("您的BMI值為:", BMI)
        if BMI < 18.5:
           print("體重過輕")
        elif 18.5 <= BMI < 24:
           print("體重正常")
        elif 24 <= BMI < 27:
           print("體重過重")
        elif 27 <= BMI < 30:
            print("輕度肥胖")
        elif 30 <= BMI < 35:
            print("中度肥胖")
        else:
            print("重度肥胖")
finally:    print("程式結束")
```
