```python
'''
有一個names.txt

讀取names.txt

with open("names.txt",encoding="utf-8") as file:

content:str = file.read()

names:list = content.split("\n")for name in names:

print(name)



請隨機取出3個名字
'''
import random

def select_random_names(filename: str, count: int) -> list:
    """
    從指定的文件中隨機選取指定數量的名字。
    
    Args:
        filename (str): 包含名字的文件路徑
        count (int): 要選取的名字數量   
        
    Returns:
        list: 隨機選取的名字列表
    """
    with open(filename, encoding="utf-8") as file:
        content: str = file.read()
        names: list = content.split()
        selected_names = random.sample(names, count)
        return selected_names

if __name__ == "__main__":
    name_count=int(input("請輸入要取出的名字數量: "))
    random_names = select_random_names("names.txt", name_count)
    for name in random_names:
        print(name)
```
