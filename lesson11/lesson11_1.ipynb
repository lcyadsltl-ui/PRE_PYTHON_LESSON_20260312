```python
'''
有一個names.txt

讀取names.txt

with open("names.txt",encoding="utf-8") as file:

content:str = file.read()

names:list = content.split("\n")for name in names:

print(name)



請隨機取出3個名字
'''
import random

def select_random_names(filename: str, count: int) -> list:
    """
    從指定的文件中隨機選取指定數量的名字。
    
    Args:
        filename (str): 包含名字的文件路徑
        count (int): 要選取的名字數量   
        
    Returns:
        list: 隨機選取的名字列表
    """
    with open(filename, encoding="utf-8") as file:
        content: str = file.read()
        names: list = content.split()
        selected_names = random.sample(names, count)
        return selected_names



def generate_student_data(names: list) -> list:
    """
    接收姓名列表，並為每個人產生 3 個 (50~100) 的隨機分數。
    回傳一個 2 維列表。
    """
    students_data = []  # 初始化容器
    
    for name in names:
        # 使用列表推導式產生 3 個隨機分數
        scores = [random.randint(50, 100) for _ in range(3)]
        
        # 將姓名與分數組合成子列表，並加入主容器
        student_record = [name] + scores
        students_data.append(student_record)
    
    return students_data

if __name__ == "__main__":
    name_count=int(input("請輸入要取出的名字數量: "))
    random_names = select_random_names("names.txt", name_count)
    results = generate_student_data(random_names)
    for student in results:
        print(student)
```
