```python
def find_factors():
    print("正在執行找因數的功能...")   
    try:
        # 提示使用者輸入一個整數
        num = int(input("請輸入一個整數："))
        
        # 處理 0 的特殊情況（0 有無限多個因數，通常不討論）
        if num == 0:
            print("0 有無限多個因數。")
            return

        # 取絕對值，確保負數也能正確計算因數
        target = abs(num)
        factors = []

        # 迴圈從 1 執行到該數字本身
        for i in range(1, target + 1):
            if target % i == 0:
                factors.append(i)
        
        # 如果原始輸入是負數，通常因數也包含負值，這裡列出正因數作為代表
        print(f"數字 {num} 的所有正因數為：{factors}")
        print(f"共有 {len(factors)} 個因數。")

    except ValueError:
        print("輸入錯誤！請確保輸入的是一個整數。")
print("我是 math_tool.py，我被載入了！")

# 執行程式
if __name__ == "__main__":
    find_factors()
```

```python
def find_common_factors():
    try:
        # 輸入兩個整數
        num1 = int(input("請輸入第一個整數："))
        num2 = int(input("請輸入第二個整數："))

        # 取絕對值，確保負數也能計算
        n1, n2 = abs(num1), abs(num2)

        # 找出較小的數，因為公因數不可能大於較小的那個數
        smaller = min(n1, n2)
        
        common_factors = []

        # 處理 0 的特殊情況
        if n1 == 0 or n2 == 0:
            print("其中一個數字為 0，公因數討論通常以非零整數為主。")
            return

        # 只需要檢查到較小的那個數即可
        for i in range(1, smaller + 1):
            if (n1 % i == 0) and (n2 % i == 0):
                common_factors.append(i)

        print(f"\n數字 {num1} 與 {num2} 的公因數有：")
        print(common_factors)
        print(f"最大公因數 (GCD) 是：{common_factors[-1]}")

    except ValueError:
        print("輸入錯誤！請確保輸入的是整數。")

if __name__ == "__main__":
    find_common_factors()
```
